```python
import numpy as np
import matplotlib.pyplot as plt
import os
import fnmatch
import yaml
import scipy

import gc

gc.collect()
```

```python
folder_path="../figure7/secundary_structure_data/"

structures={}
for j,file in enumerate(os.listdir(folder_path)):
    if fnmatch.fnmatch(os.fsdecode(file), "*.dat"):
        structures[os.fsdecode(file)[:-12]]=np.loadtxt(folder_path+os.fsdecode(file))
```

```python
weights={'H':1,'C':12,'N':14,'S':32,'O':16}

fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')
plt.rcParams["figure.figsize"] = [6.00, 6]
plt.rcParams.update({'font.size': 10})
plt.rcParams['figure.autolayout']=True
    
fig, axes = plt.subplots(3,1, gridspec_kw={'height_ratios': [5, 5,5]})
fig.patch.set_facecolor('xkcd:white')
(ax1, ax2,ax3)=axes.flatten()

WEIG=[]
T1=[]
T2=[]
NOE=[]

colors={'eElab':0,'eYqjD':1,'yFis1':2,'hMff':3,'mag2':4,'GWALP':5}

simulation_path='/media/ricky/Ricky2020/Corys_peptides/full_trajectories/'
for name in os.listdir(path_yaml):
    if '_relax.yaml' in name:
        prefix=name[:-11]
        if prefix in os.listdir(simulation_path):
            path=f'{simulation_path}{prefix}/'
            r_path=f'{path}/README.yaml'
            if os.path.isfile(r_path):
                with open(r_path) as f:
                    readme=yaml.safe_load(f)
                    weight=0
                    with open(f"{path}{readme['FILES_FOR_RELAXATION']['gro']['NAME']}",'r') as f:
                        for line in f:
                            if len(line.split())==6:
                                if line.split()[1]!='SOD':
                                    weight+=(weights[line.split()[1][0]])
                #print(prefix,weight)
                with open(path_yaml+name) as yaml_file:
                    yamlFile = yaml.load(yaml_file, Loader=yaml.FullLoader)

                yamlFile=yamlFile["analysis0"]
                
                
                T1_val=[]
                T2_val=[]
                NOE_val=[]
                for i in yamlFile["results"]:
                    T1_val.append(yamlFile["results"][i]["T1"])
                    T2_val.append(yamlFile["results"][i]["T2"])
                    NOE_val.append(yamlFile["results"][i]["hetNOE"])
                    
                
                for peptide,col in colors.items():
                    if peptide in prefix:
                        color=f'C{col}'
                        break
                ax1.plot([weight]*len(T1_val),T1_val,'o',c=color)
                ax2.plot([weight]*len(T2_val),T2_val,'o',c=color)
                ax3.plot([weight]*len(NOE_val),NOE_val,'o',c=color)
            
                
                T1+=T1_val
                T2+=T2_val
                NOE+=NOE_val
                WEIG+=[weight]*len(T1_val)

coeff = np.polyfit(WEIG, T1,1,full=True)
xn = np.linspace(min(WEIG),max(WEIG),100)
yn = np.poly1d(coeff[0])
r, p = scipy.stats.pearsonr(WEIG, T1)
if p < 0.001:
    p=" < 0.001"
else:
    p=" = "+str(np.round(p,3))
ax1.plot(xn,yn(xn),"-",label="corr. coef. = "+str(np.round(r,2))+", p-value"+p)

ax1.legend()

#r = np.corrcoef(hel, T1)

coeff = np.polyfit(WEIG, T2,1,full=True)
yn = np.poly1d(coeff[0])
r, p = scipy.stats.pearsonr(WEIG, T2)
if p < 0.001:
    p=" < 0.001"
else:
    p=" = "+str(np.round(p,3))
ax2.plot(xn,yn(xn),"-",label="corr. coef. = "+str(np.round(r,2))+", p-value"+p)

ax2.legend()

coeff = np.polyfit(WEIG, NOE,1,full=True)
yn = np.poly1d(coeff[0])
r, p = scipy.stats.pearsonr(WEIG, NOE)
if p < 0.001:
    p=" < 0.001"
else:
    p=" = "+str(np.round(p,3))
ax3.plot(xn,yn(xn),"-",label="corr. coef. = "+str(np.round(r,2))+", p-value"+p)
ax3.legend(loc=4)
    
    
ax1.set_title("Effect of moleculear weight on T1, T2 and hetNOE values ") 

ax3.set_xlabel("Molecular weight [u]")
ax1.set_ylabel("T1 [s]")
ax2.set_ylabel("T2 [s]")
ax3.set_ylabel("hetNOE")
plt.savefig('weight_vs_spin_relax.pdf')
```

```python
path_yaml="../../relax_times_yamls/protein/raw_data/"
skip=1
cutoff=1

weig=1
ms=2

xn=np.linspace(0,1,100)

fig = plt.figure()
fig.patch.set_facecolor('xkcd:white')
plt.rcParams["figure.figsize"] = [3.00, 3.5]
plt.rcParams.update({'font.size': 5})
    
fig, axes = plt.subplots(4,1, gridspec_kw={'height_ratios': [5, 5,5,1]})
fig.patch.set_facecolor('xkcd:white')
(ax1, ax2,ax3,ax4)=axes.flatten()

#store all but skip values
hel=[]
T1=[]
T2=[]
NOE=[]

#store skiped values
hel_end=[]
T1_end=[]
T2_end=[]
NOE_end=[]


#store values close to helical relins
hel_close=[]
T1_close=[]
T2_close=[]
NOE_close=[]

color_map=["#2b1100","#552200","#803300","#aa4400","#d45500","#ff6600","#ff7f2a","#ff9955","#ffb380","#ffccaa","#ffe6d5"]

for i,name_pre in enumerate(structures):
    try:
        for j,file in enumerate(os.listdir(path_yaml)):
            if fnmatch.fnmatch(os.fsdecode(file), "*"+name_pre+"*.yaml"):

                with open(path_yaml+os.fsdecode(file)) as yaml_file:
                    yamlFile = yaml.load(yaml_file, Loader=yaml.FullLoader)

        yamlFile=yamlFile["analysis0"]

        name=structures[name_pre]

        if "GWALP" in name_pre:
            support_table=name[0]
            support_table=support_table.tolist()
        elif "dimer" in name_pre:
            lent=len(name[0])
            support_table=name[0][1:lent//2].tolist()+name[0][lent//2+1:].tolist()

        else:
            support_table=name[0][1:]
            support_table=support_table.tolist()


        T1_val=[]
        T2_val=[]
        NOE_val=[]
        for i in yamlFile["results"]:
            T1_val.append(yamlFile["results"][i]["T1"])
            T2_val.append(yamlFile["results"][i]["T2"])
            NOE_val.append(yamlFile["results"][i]["hetNOE"])
            
        
        if not "yFis1_micelle_50SDS_CHARMM_310K_Na_Neut_OPC_replica2" in name_pre:

            
            for i,helicity in enumerate(support_table):
                if helicity<cutoff:
                    if i>0 and i<len(support_table)-2:
                        neig_value=helicity
                        colorit=color_map[int(np.round(neig_value*10,0))]
                        
                        helicity=(support_table[i-1]+support_table[i+1]+weig*helicity)/(2+weig)
                        support_table[i]=helicity
                        ax1.plot(helicity,T1_val[i],"o",c=colorit,markersize=ms)
        
                        ax2.plot(helicity,T2_val[i],"o",c=colorit,markersize=ms)
                        ax3.plot(helicity,NOE_val[i],"o",c=colorit,markersize=ms)
                    elif i==0:
                        neig_value=helicity
                        colorit=color_map[int(np.round(neig_value*10,0))]
                        
                        helicity=(support_table[i+1]+weig*helicity)/(1+weig)
                        support_table[i]=helicity
                        ax1.plot(helicity,T1_val[i],"o",c=colorit,markersize=ms)
        
                        ax2.plot(helicity,T2_val[i],"o",c=colorit,markersize=ms)
                        ax3.plot(helicity,NOE_val[i],"o",c=colorit,markersize=ms)
                    else:
                        neig_value=helicity
                        colorit=color_map[int(np.round(neig_value*10,0))]
                        
                        helicity=(support_table[i-1]+weig*helicity)/(1+weig)
                        support_table[i]=helicity
                        ax1.plot(helicity,T1_val[i],"o",c=colorit,markersize=ms)
        
                        ax2.plot(helicity,T2_val[i],"o",c=colorit,markersize=ms)
                        ax3.plot(helicity,NOE_val[i],"o",c=colorit,markersize=ms)

        
        
            support_table=support_table[skip:-skip]
            T1_val=T1_val[skip:-skip]
            T2_val=T2_val[skip:-skip]
            NOE_val=NOE_val[skip:-skip]


            hel=hel+support_table
            T1=T1+T1_val
            T2=T2+T2_val
            NOE=NOE+NOE_val


        
        
                
    except Exception as e: 
        print(e, name_pre)
        #print((yamlFile["results"]))
        pass

    
coeff = np.polyfit(hel, T1,1,full=True)
xn = np.linspace(0,1,100)
yn = np.poly1d(coeff[0])
r, p = scipy.stats.pearsonr(hel, T1)
if p < 0.001:
    p=" < 0.001"
else:
    p=" = "+str(np.round(p,3))
ax1.plot(xn,yn(xn),"-",label="corr. coef. = "+str(np.round(r,2))+", p-value"+p)

ax1.legend()

#r = np.corrcoef(hel, T1)

coeff = np.polyfit(hel, T2,1,full=True)
yn = np.poly1d(coeff[0])
r, p = scipy.stats.pearsonr(hel, T2)
if p < 0.001:
    p=" < 0.001"
else:
    p=" = "+str(np.round(p,3))
ax2.plot(xn,yn(xn),"-",label="corr. coef. = "+str(np.round(r,2))+", p-value"+p)

ax2.legend()

coeff = np.polyfit(hel, NOE,1,full=True)
yn = np.poly1d(coeff[0])
r, p = scipy.stats.pearsonr(hel, NOE)
if p < 0.001:
    p=" < 0.001"
else:
    p=" = "+str(np.round(p,3))
ax3.plot(xn,yn(xn),"-",label="corr. coef. = "+str(np.round(r,2))+", p-value"+p)

    
    
ax1.set_title("Effect of helicity on T1, T2 and hetNOE values ") 

ax3.set_xlabel("Local Environment Helicity [%/100]")
ax1.set_ylabel("T1 [s]")
ax2.set_ylabel("T2 [s]")
ax3.set_ylabel("hetNOE")
ax1.legend(loc=1)
ax2.legend()
ax3.legend(loc=4)

#cax = plt.axes([0.85, 0.1, 0.075, 0.8])
#plt.colorbar()

cmap = plt.cm.get_cmap(newcmp)
colors = cmap(np.arange(cmap.N))

#ax4([0.08, 0.08, 0.94-0.08, 0.94-0.08])
ax4.imshow([colors], extent=[0, 1, 0, 0.01])
ax4.set_yticks([])
ax4.set_xlabel("Helicity of given residue")
plt.savefig("neighbours.pdf")    
plt.show()
```
