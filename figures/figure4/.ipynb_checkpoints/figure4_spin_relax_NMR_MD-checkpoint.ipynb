```python
OP=0 # order parameter
smallest_corr_time=0 # enter in log scale -3 fs; 0 ps; 3 ns; 6 us;
biggest_corr_time=5 # same as above
N_exp_to_fit=100 # number of exponential functions to be fitted between the samlles and biggest corr time
analyze=1/50 # the proportin of correlation data to be used for fitting, ex. 1/2 uses first half of the data
magnetic_fields=[2.35,5.99,8.49] #2.35 # 5.99 # 8.49 T (values used in SDS paper, J.Chem. Soc.,, Faraday Trans. 1, 1988, 84(12), 4475-4486)
magn_field_MHz=850 # for values in MHz
magnetic_field=magn_field_MHz*2*np.pi/gammaH*10**6
#magnetic_field=8.49
nuclei="15N" #nuclei to calculate: 2H-deutherium; 13C - carbon; 15N - nitrogen 


relax_data={}
for simulation in relaxation_yamls:

    pituus=0
    for ana in timescales_yamls[simulation]:
        info=timescales_yamls[simulation][ana]["info"]
        if (info['03_N_exp_to_fit']==N_exp_to_fit and
           info['04_smallest_corr_time_[s]']==10**(smallest_corr_time-12)
            and info['05_biggest_corr_time_[s]']==10**(biggest_corr_time-12)
           and info["06_analyze"]==analyze):
            if info['08_corr_func_length_[ps]']>pituus:
                analysis=ana
            pituus=max(pituus,info['08_corr_func_length_[ps]'])

        pituus=0
        for ana in relaxation_yamls[simulation]:
            info=relaxation_yamls[simulation][ana]["info"]
            if (info['03_N_exp_to_fit']==N_exp_to_fit and
               info['04_smallest_corr_time_[s]']==10**(smallest_corr_time-12)
               and info['05_biggest_corr_time_[s]']==10**(biggest_corr_time-12)
               and info["06_analyze"]==analyze and
               info["01_magnetic_field_[T]"]==magnetic_field):
                if info['08_corr_func_length_[ps]']>pituus:
                    analysis_relax=ana
                pituus=max(pituus,info['08_corr_func_length_[ps]'])        
                #print(pituus)



        relax_data[simulation]={}
        relax_data[simulation]["results"]={}
        relax_data[simulation]["info"]=timescales_yamls[simulation][analysis]["info"]
        relax_data[simulation]["info"]["07_OP"]=OP
        relax_data[simulation]["info"]["01_magnetic_field_[T]"]=magnetic_field
        relax_data[simulation]["info"]["02_magnetic_field_[MHz]"]=float(np.round(magnetic_field /(2*np.pi/gammaH*10**6),2))
        relax_data[simulation]["info"]["00_nuclei"]=nuclei



        for residue in timescales_yamls[simulation][analysis]["results"]["Coeff"]:
            #bigTS=1000

            if timescales_yamls[simulation][analysis]["results"]["Coeff"][residue][-1]>0:
                timescales_yamls[simulation][analysis]["results"]["Coeff"][residue][-1]=0

            T1, T2, NOE = rt.choose_nuclei[nuclei](magnetic_field,timescales_yamls[simulation][analysis]["results"]["Coeff"][residue],timescales_yamls[simulation][analysis]["results"]["Ctime"],OP)
            relax_data[simulation]["results"][residue]={} 
            relax_data[simulation]["results"][residue]["T1"]=float(T1)
            relax_data[simulation]["results"][residue]["T2"]=float(T2)
            relax_data[simulation]["results"][residue]["hetNOE"]=float(NOE)

            relaxation_yamls[simulation][analysis_relax]["results"][residue]["T1"]=float(T1)
            relaxation_yamls[simulation][analysis_relax]["results"][residue]["T2"]=float(T2)
            relaxation_yamls[simulation][analysis_relax]["results"][residue]["hetNOE"]=float(NOE)
```
