```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
radius_hmff=np.loadtxt("radius_hmff.dat")
radius_gwalp=np.loadtxt("radius_gwalp.dat")
radius=np.loadtxt("radius_water.dat")
timescales_hmff=np.loadtxt("timescale_hmff.dat")
timescales_gwalp=np.loadtxt("timescale_gwalp.dat")
real_water=np.loadtxt("timescale_real_water.dat")
```

```python
plt.rcParams["figure.figsize"] = [3, 1.5]
plt.rcParams["figure.autolayout"] = True
plt.rcParams.update({'font.size': 5})


plt.plot(radius_hmff,timescales_hmff,"o",label='hMff monomer',c="C3")
plt.plot(radius_gwalp,timescales_gwalp,"o",label='GWALP dimer',c="C5")

coeff = np.polyfit(list(radius_gwalp)+list(radius_hmff), list(timescales_gwalp)+list(timescales_hmff),1,full=True)
yn = np.poly1d(coeff[0])
xn = [1/(1.8**3),1/(1.45**3)]


plt.plot(radius,real_water,label="A perfect solid sphere in water",c='black')
#plt.plot(radius,real_water2,label="A perfect solid sphere in water")

plt.plot(xn,yn(xn),"--",c="C3")
xn = [1/(1.79**3),1/(1.45**3)]
plt.plot(xn,yn(xn),"--",c="C5")
plt.legend(loc=2)
#plt.yscale('log')
plt.xlabel("1/R$^3$ [nm$^{-3}$]")
plt.ylabel("1/Ï„  [ns$^{-1}$]")
plt.text(0.262,0.23,"1/t = 0.43 * 1/R$^3$ + 0.05")
plt.text(0.28,0.37,"1/t = 4.64 * 1/R$^3$")
#plt.title("Average of the most abundand timescale over AA vs micelle radius")

plt.savefig("rot_vs_rad2.pdf")
```
